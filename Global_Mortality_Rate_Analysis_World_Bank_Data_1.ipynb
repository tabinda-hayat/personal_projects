```r
# train_test on imputed data 
imputed_data <- imputed_dataset_asia
set.seed(100)
train_idx = sample(1:nrow(imputed_data), 0.6*nrow(imputed_data))
train_data = imputed_data[train_idx, -c(1,2,11)]
test_data = imputed_data[-train_idx, -c(1,2,11)]

train_y = imputed_data$Mortality_rate[train_idx]
test_y = imputed_data$Mortality_rate[-train_idx]

dim(train_data); dim(test_data)
```

# Regression Analysis

```r
# reg_model <- lm(Mortality_rate ~., data = train_data)
# model_summary <- summary(reg_model)
# model_summary


reg_model <- lm(Mortality_rate~., data = train_data)
model_summary <- summary(reg_model)
model_summary
```

```r
pred_y <- predict.lm(reg_model, newdata=test_data)
sqrt(mean((pred_y-test_y)**2))
```

```r
library(glmnet)


#Setting the train data and tran response as matrix and a vector respectively 
ridge_train_x =model.matrix (Mortality_rate~.,train_data)
ridge_train_y =train_data$Mortality_rate

#Setting the test data and tran response as matrix and a vector respectively 
ridge_test_x=model.matrix (Mortality_rate~.,test_data)
ridge_test_y=test_data$Mortality_rate


# #Using cross-validation to the optimal lambda in ridge regression 
set.seed (100)
ridge_cv=cv.glmnet(ridge_train_x,ridge_train_y,alpha =0)

best_lambda <- ridge_cv$lambda.min
paste("The best lambda value is:", best_lambda)

options(repr.plot.width= 8, repr.plot.height= 8)
plot(ridge_cv)


# #Setting the train data and tran response as matrix and a vector respectively 
# ridge_train_x =model.matrix (Mortality_rate_under_5_per_1000~.,train_data)
# ridge_train_y =train_data$Mortality_rate_under_5_per_1000

# #Setting the test data and tran response as matrix and a vector respectively 
# ridge_test_x=model.matrix (Mortality_rate_under_5_per_1000~.,test_data)
# ridge_test_y=test_data$Mortality_rate_under_5_per_1000


# # #Using cross-validation to the optimal lambda in ridge regression 
# set.seed (100)
# ridge_cv=cv.glmnet(ridge_train_x,ridge_train_y,alpha =0)

# best_lambda <- ridge_cv$lambda.min
# paste("The best lambda value is:", best_lambda)

# options(repr.plot.width= 8, repr.plot.height= 8)
# plot(ridge_cv)
```

```r
#Training the model with optimal lambda value
best_ridge_model =glmnet(ridge_train_x,ridge_train_y,alpha=0,lambda=best_lambda)
# best_ridge_model$beta

#Retrieving the ridge coefficients in ridge model 
ridge_coef=predict(best_ridge_model,type="coefficients",s=best_lambda)[1:length(best_ridge_model$beta),]

# #Non zero coefficients
as.data.frame(ridge_coef[ridge_coef!=0])
```

```r
# Preiction on test data 
best_ridge_pred=predict(best_ridge_model,s=best_lambda ,newx=ridge_test_x)
#Calculating the Mean Squared Error
rmse=sqrt(mean((best_ridge_pred-ridge_test_y)^2))

cat("The Mean Squared Error (RMSE) for the prediction on test set: ", rmse)
```

```r
ridge_coef = ridge_coef[ridge_coef!=0][-c(1)]
ridge_coef = data.frame("Indicators"=c("Health expenditure", "Out-pocket expenditure", 
"Education expenditure", "Number of physicians", "Female literacy", "Open defecation"), "Value" = ridge_coef )
ridge_coef =  data.frame("Indicators"=ridge_coef$Indicators, "Value"=ridge_coef$Value)
ridge_coef
```

```r
library(ggplot2)
color <- ifelse(ridge_coef$Value < 0, "blue", "red")
ggplot(ridge_coef, aes(x=Indicators, y=Value))+
  geom_bar(stat = "identity",
           show.legend = FALSE,
           fill = color) + 
             geom_text(aes(label = round(Value,2), vjust = ifelse(Value >= 0, 0, 1)), size=6.5)+
  xlab(" ") +
  ylab("Value") +
  lims(y=c(-2.5, 2.5))+theme(axis.title.x = element_text(size = 30), 
                        axis.title.y = element_text(size = 25), text = element_text(size = 30,angle = 90))
                        
```

```r
#Best Regression Predictions
options(repr.plot.width= 8, repr.plot.height= 8)
plot(test_y, best_ridge_pred, xlab="Actual mortality rate", 
ylab="Predicted mortality rate", main= paste("RMSE: " ,round(rmse,2), " and R^2: ", round(best_ridge_model$dev.ratio,2) ),
pch=5, col = "red", font=50, cex.lab = 1.6,
     cex.axis = 1.6,
     cex.main = 1.6,
     cex.sub = 1.6)
# lines( seq(0,max(test_y), length = 30), seq(0,max(best_lasso_pred), length = 30))
```

```r
######## Lasso

#Using cross-validation to choose the tuning parameter ??

lasso_train_x <- ridge_train_x 
lasso_train_y <- ridge_train_y 

#Using cross-validation to the optimal lambda in lasso regression 
set.seed (100)
lasso_cv=cv.glmnet(lasso_train_x,lasso_train_y,alpha =1)

options(repr.plot.width= 8, repr.plot.height= 8)
plot(lasso_cv)
```

```r
best_lambda_ls <- lasso_cv$lambda.min
paste("The best lambda value for lasso regression is:", best_lambda_ls)


#Training the lasso model with optimal lambda value 
best_lasso_model =glmnet(lasso_train_x,lasso_train_y,alpha=1,lambda=best_lambda_ls)
# best_lasso_model$beta; 
paste(" The R^2 value:", best_lasso_model$dev.ratio)
```

```r
#Retrieving the lasso coefficients in lasso model 
lasso_coef=predict(best_lasso_model,type="coefficients",s=best_lambda_ls)[1:length(best_lasso_model$beta),]

#Non zero coefficients
as.data.frame(lasso_coef[lasso_coef!=0])
```

```r
lasso_test_x <- ridge_test_x
lasso_test_y <- ridge_test_y

# Preiction on test data 
best_lasso_pred=predict(best_lasso_model,s=best_lambda_ls ,newx=lasso_test_x)
#Calculating the Mean Squared Error
rmse=sqrt(mean((best_lasso_pred-lasso_test_y)^2))

cat("The Mean Squared Error (RMSE) for the prediction on test set in Lasso model: ", rmse)
```
